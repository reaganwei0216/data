```python
# --- 第 1 部分 ---
# 載入函式庫與資料集
import openensembles as oe
import numpy as np
import pandas as pd
import sklearn.metrics

from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()

np.random.seed(123456)
```

```python
bc
```

```python
bc.data
```

```python
pd.DataFrame(bc.data)
```

```python
# --- 第 2 部分 ---
# 建立資料
cluster_data = oe.data(pd.DataFrame(bc.data), bc.feature_names)
```

```python
cluster_data
```

```python
ensemble = oe.cluster(cluster_data)
help(ensemble)
```

# 原始資料的集成分群

```python
# --- 第 3 部分 ---
# 訓練模型並計算同質性
for K in [2, 3, 4, 5, 6, 7]:             # 不同子群數 
    for ensemble_size in [3, 4, 5]:
        ensemble = oe.cluster(cluster_data)
        for i in range(ensemble_size):       # 不同基學習器數量       
            name = f'kmeans_{ensemble_size}_{i}'
            ensemble.cluster('parent', 'kmeans', name, K)

        preds = ensemble.finish_majority_vote(threshold=0.5)  # threshold參數決定多少比例的基學習器有一致的分類便將資料視為一群
        print(f'K: {K}, size {ensemble_size}:', end=' ')
        print('%.2f' % sklearn.metrics.homogeneity_score(
                bc.target, preds.labels['majority_vote']))
```

# 降維轉換資料的集成分群

```python
# --- 第 1 部分 ---
# 載入函式庫與資料集
import openensembles as oe
import numpy as np
import pandas as pd
import sklearn.metrics

from sklearn.datasets import load_breast_cancer
from sklearn.manifold import TSNE

bc = load_breast_cancer()

np.random.seed(123456)
```

```python
help(oe.data)
```

```python
bc.data
```

```python
t = TSNE(n_components=2)
pd.DataFrame(t.fit_transform(bc.data))
```

```python
# --- 第 2 部分 ---
# 建立資料
t = TSNE(n_components=2)
cluster_data = oe.data(pd.DataFrame(t.fit_transform(bc.data)), [0,1])
```

```python
# --- 第 3 部分 ---
# 訓練模型並計算同質性
for K in [2, 3, 4, 5, 6, 7]:
    for ensemble_size in [3, 4, 5]:
        ensemble = oe.cluster(cluster_data)
        for i in range(ensemble_size):
            name = f'kmeans_{ensemble_size}_{i}'
            ensemble.cluster('parent', 'kmeans', name, K)

        preds = ensemble.finish_majority_vote(threshold=0.5)
        print(f'K: {K}, size {ensemble_size}:', end=' ')
        print('%.2f' % sklearn.metrics.homogeneity_score(
                bc.target, preds.labels['majority_vote']))
```
