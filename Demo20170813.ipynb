### 範例檔案
- http://gcis.nat.gov.tw/mainNew/matterAction.do?method=showFile&fileNo=t70013_p

```python
from pdfminer.pdfparser import PDFParser, PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LAParams, LTTextBox, LTTextLine

s = ''
fp = open('data/t70013_p.pdf', 'rb')
parser = PDFParser(fp)
doc = PDFDocument()
parser.set_document(doc)
doc.set_parser(parser)
doc.initialize('')
rsrcmgr = PDFResourceManager()
laparams = LAParams()
device = PDFPageAggregator(rsrcmgr, laparams=laparams)
interpreter = PDFPageInterpreter(rsrcmgr, device)
# Process each page contained in the document.
for page in doc.get_pages():
    interpreter.process_page(page)
    layout = device.get_result()
    for lt_obj in layout:
        print(lt_obj)
        #pass
        #if isinstance(lt_obj, LTTextBox) or isinstance(lt_obj, LTTextLine):
        #    s = s + lt_obj.get_text()
```

```python
def pageDetail(pageurl):
    dic = {}
    res = requests.get(pageurl)
    soup = BeautifulSoup(res.text, 'html.parser')
    dic['title']   = soup.select('h1')[0].text
    dic['summary'] = soup.select('.ndArticle_margin p')[0].text
    dic['time']    = soup.select('.ndArticle_creat')[0].text[5:]
    dic['clicked'] = soup.select('.ndArticle_view')[0].text if len(soup.select('.ndArticle_view')[0]) > 0 else None 
    return dic
```

```python
newsurl = 'https://tw.appledaily.com/new/realtime/{}'
domain  = ''
newsary = []
for page in range(1,2):
    
    section = requests.get(newsurl.format(page))
    soup    = BeautifulSoup(section.text, 'html.parser')
    
    for news in soup.select('.rtddt a'):
        # 宣告空的字典dic
        dic             = {}
        # 取得連結
        dic['link']     = news['href']
        # 取得類別
        dic['category'] = news.select('h2')[0].text 
        # 取得內文資訊 回傳字典
        dic.update(pageDetail(dic['link']))
        
        newsary.append(dic)
```
