```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
randomSequences = pd.DataFrame(np.random.normal(1.0, 0.08, (100,8)))
randomSequences.head()
#randomSequences.shape
```

```python
accumulates = randomSequences.cumprod()
accumulates.head()
```

```python
%matplotlib inline
accumulates.plot(linewidth=3)
plt.title('simulation')
plt.xlabel('time sequence')
plt.ylabel('value')
plt.legend(loc=1)
```
