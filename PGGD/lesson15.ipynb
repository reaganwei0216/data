## Complete node (those node without relations)

````python
objects = ylq_all_star_relations.object.tolist()
objects[:10]
````

````python
all_star_nodes = ylq_all_star_ids.name.tolist()
other_ids = []
for obj in objects:
    if obj not in all_star_nodes:
        other_ids.append(obj)
print(other_ids)
````

````python
node_map = {}
for index, node in enumerate(all_star_nodes + other_ids):
    if node not in node_map:
        node_map[node] = index + 1
# print(node_map)
````

## 匯入輸出的csv資料，使用 neo4j 構建關係圖譜

手動去掉一些無用的列資料後，將 `./csv/ylq_all_star_nodes.csv` 和 `./csv/ylq_all_star_relations.csv` 兩個csv文件，放到

.\neo4j\import 資料夾裡，然後在開啟 `neo4j web browser` 分別執行以下兩行命令，就完成了關係圖譜的建立！如果資料量大，則可能會需要一些等待時間。

- 匯入明星 Nodes
```
LOAD CSV WITH HEADERS FROM 'file:///ylq_all_star_nodes.csv' AS data CREATE (:star{starname:data.name, starid:data.id});
```

- 匯入明星關係 Relationships
```
LOAD CSV WITH HEADERS FROM "file:///ylq_all_star_relations.csv" AS relations
MATCH (entity1:star{starname:relations.subject}) , (entity2:star{starname:relations.object})
CREATE (entity1)-[:rel{relation: relations.relation}]->(entity2)
```

## Cypher

- 查詢某位藝人
```python
MATCH (n:star{starname:'那英'})-[r]-(m) RETURN * LIMIT 25
```
```python
MATCH (p:star) WHERE p.starname =~ '.*梁朝偉.*' RETURN p
```

- 查詢某位藝人的交友關係
```python
MATCH (p:star)-[]-(m) WHERE p.starname =~ '.*梁朝偉.*' RETURN *
```

- 查詢某位藝人的好友們
```python
MATCH (n:star{starname:'那英'})-[r:rel{relation:'好友'}]-(m) RETURN * LIMIT 25
```

- 查詢關係
```python
MATCH p=()-[r:rel]->() RETURN p LIMIT 25
```

- 查詢所有為密友關係的明星
```python
MATCH p=()-[r:rel{relation:"密友"}]->() RETURN p LIMIT 25
```

- 查詢怨偶
```python
MATCH p=()-[r:rel{relation:'前任夫妻'}]->() RETURN p LIMIT 25
```

- 查詢所有藝人是好友的關係
```python
MATCH (n:star)-[r:rel{relation:"好友"}]-(m) RETURN * LIMIT 25
```

- 查詢某位藝人的某個好友
```python
MATCH (star:star {starname:"那英"})-[:rel{relation:"好友"}]->(starsFriends)
WHERE starsFriends.starname="李亞鵬"
RETURN star, starsFriends
LIMIT 10
```

- 查詢那英的關係中，其他明星與王菲關係中為好友關係的人
```python
MATCH (star:star {starname:"那英"})-[:rel]->(m)<-[:rel{relation:"好友"}]-(coFriends)
RETURN coFriends
```

- 查詢王菲的關係中，其他明星與王菲關係中為密友關係的人
```python
MATCH (star:star {starname:"王菲"})-[:rel]->(m)<-[:rel{relation:"密友"}]-(coFriends)
RETURN coFriends, star
```

- 查詢查詢王菲的關係中，其他明星與王菲關係中為舊愛關係的人
```python
MATCH (star:star {starname:"王菲"})-[:rel]->(m)<-[:rel{relation:"舊愛"}]-(coFriends)
RETURN coFriends, m, star
```
