## 如何讓兩個list 相乘

```python
a = [1,3,5,7,9]
b = [2,4,6,8,10]

# method 1
for i in range(0, len(a)):
    print(a[i] * b[i])

# method 2
ary = []
for i in range(0, len(a)):
    ary.append(a[i] * b[i])
print(ary)

# method 3
print([a[i] * b[i] for i in range(0, len(a))])

# method 4: zip
for f, s in zip(a, b):
    print(f * s)

# method 5
print([f * s for f,s in zip(a,b)])
```

```python
#  zip
a = [1,3,5,7,9]
b = [2,4,6,8,10]
print(a)
print(b)
print(zip(a, b))
for item in zip(a, b):
    print (item)
for f,s in zip(a, b):
    print (f*s)
```

### 比對數字

```python
a = '5'
# [] 理面的元素是用"或" 來比對
print(re.match('[0123456789]',a))

# - 代表連續
print(re.match('[0-9]',a))

# \d == [0-9]
print(re.match('\d',a))
```

### 比對英文字母

```python
s = 'g'
print(re.match('[abcdefghijklmnopqrstuvwxyz]', s))

# - 代表連續
print(re.match('[a-z]', s))

s2 = 'G'
# 比對大寫
print(re.match('[a-z]', s2)) # 只能比對到小寫

print(re.match('[a-zA-Z]', s2))
```

### 同時比對數字跟英文字

```python
s = '5'
re.match('[a-zA-Z0-9]', s)

# \w == [a-zA-Z0-9]
re.match('\w', s)
```

### 比對特殊符號

```python
a = '$'
re.match('\w', a)

# . == 任意字元 
re.match('.', a)
```

### 比對車牌

```python
s = '9988'

# {} 理面的數字代表要比對到的數量
re.match('\d{4}', s)

# {s,e} s 代表至少有多少符號, e 代表至多有多少符號
re.match('\d{3,5}', s)

# {s, } 至少比對到 s 數量符號
re.match('\d{3,}', s)

# {0, } 至少比對到 0 數量以上符號
re.match('\d{0,}', s)

# * == {0, } 
re.match('\d*', s)


# {1, } 至少比對到 1 數量以上符號
re.match('\d{1,}', s)

# + == {1, } 
re.match('\d+', s)
```

## 正規表達法實際案例

```python
# + == {1, } 
import re
email = 'david@largidata.com'
m = re.match('(\w+)@(\w+)', email)
print(m.groups())
print(type(m.groups()))

m = re.match('(\w+)@([\w.]+)', email)
print(m.groups())
print(m.group(0))
print(m.group(1))
print(m.group(2))
```

```python
n = '1999.5'
m = re.match('(\d+)\.(\d+)',n)
print(m.groups())
print(m.groups(0))
```

```python
name = 'David Chiu'
m = re.match('(\w+) (\w+)', name)
print(m.group(1))
print(m.group(2))

m = re.match('(?P<first_name>\w+) (?P<last_name>\w+)', name)
print(m.group('first_name'))
print(m.group('last_name'))

command = 'scp 1.txt root@10.0.0.1:./'
m = re.match('^scp (.+?) (\w+)@([\d.]+):(.+)', command)
m.groups()
```
